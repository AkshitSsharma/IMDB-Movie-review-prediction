```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

# Statistical ML Approach

**Step 1**
- Gathering the data

```python
df=pd.read_csv("IMDB Dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

**Changing sentiment into a number**

```python
sentiments_1=df["sentiment"].unique()
sentiments_1
```

```python
sentiment_number={}
i=0
for senti in sentiments_1:
    sentiment_number[senti]=i
    i+=1
df["sentiment"]=df["sentiment"].map(sentiment_number)    
```

**Step 2**
- Text Cleaning

*Checking Lowercase*

```python
df["review"]=df["review"].apply(lambda x:x.lower())
```

*Checking Punctuation*

```python
import string
def remove_punc(txt):
    return txt.translate(str.maketrans("","",string.punctuation))
```

```python
df["review"]=df["review"].apply(remove_punc)
```

*Removing Numbers*

```python
def remove_numbers(txt):
    new=""
    for i in txt:
        if not i.isdigit():
            new+=i
    return new 
```

```python
df["review"]=df["review"].apply(remove_numbers)
```

*Remove html tags*

```python
import re

def remove_html_tags(txt):
    return re.sub(r'<.*?>', '', txt)

df["review"] = df["review"].apply(remove_html_tags)
```

*Removing URL/Links*

```python
def remove_urls(txt):
    return re.sub(r'http\S+|www\S+|https\S+', '', txt, flags=re.MULTILINE)
df["review"] = df["review"].apply(remove_urls)
```

*Removing Emojis and Special Characters*

```python
def remove_emojis(txt):
    new=""
    for i in txt:
        if i.isascii():
            new+=i
    return new
df["review"]=df["review"].apply(remove_emojis)     
```

*Removing Stopwords*

```python
#Natural Language Toolkit
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

```python
stop_words=set(stopwords.words("english"))
```

```python
def remove(txt):
    word=txt.split()
    cleaned=[]
    for i in word:
        if not i in stop_words:
            cleaned.append(i)
    return ' '.join(cleaned)     
```

```python
df["review"]=df["review"].apply(remove)
```

```python
df["review"].loc[0]
```

# Splitting Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df["review"]      
y = df["sentiment"]   
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# Feature Extraction 
**TF-IDF(Term frequency-Inverse Document frequency)**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer()
```

```python
X_train_tf=vectorizer.fit_transform(X_train)
X_test_tf=vectorizer.transform(X_test)
```

# Machine Learning Models for Sentiment Classification

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model_LR=LogisticRegression()
```

```python
model_LR.fit(X_train_tf,y_train)
```

```python
pred_LR=model_LR.predict(X_test_tf)
print("Logistic Regression Accuracy:", round(accuracy_score(y_test, pred_LR), 3))
```

**Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model_NB=MultinomialNB()
```

```python
model_NB.fit(X_train_tf,y_train)
```

```python
pred_NB=model_NB.predict(X_test_tf)
```

```python
print("Naive Bayes Accuracy:", round(accuracy_score(y_test, pred_NB), 3))
```
